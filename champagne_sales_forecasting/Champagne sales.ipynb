# Time series forecasting (using Champagne sales data)

This notebook can be seen as a *time series forecasting demo* using Champagne sales data I obtained from an Udacity online course. The goal is to demonstrate data preparation (transforming, detrending and removing seasonal components) and application of a time series forecasting model. The data preparation is needed to make the time series stationary. This is typically the requirement for applying forecasting models, such as the Seasonal ARIMA (= SARIMA) model that I am going to use subsequently.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import statsmodels.api as sm
```

First, load the data into a dataframe, then transform the month column to turn it into a DateTimeIndex.

```python
df = pd.read_excel("champagne-sales.xlsx", header=0, names=['month', 'sales'])
df.head()
```

```python
df.month = df.month.apply(lambda x: pd.to_datetime('200' + x, format='%Y-%m'))
df.set_index('month', inplace=True)
df.info()
```

## Stationarity

To build a forecasting model for a time series, it is important to transform the series into a (near-)stationary time series, i.e. with consistent mean and standard variation. This process might include several steps. I will start by preparing a function that will make it easier later on to perform repeated stationarity checks.

```python
from statsmodels.tsa.stattools import adfuller

def validate_stationarity(series):
    """ This function plots a time series including rolling mean and standard deviation. 
    It also performs the Dickey-Fuller Test, a statistical test to check for stationarity. """
    
    plt.figure(figsize=(12,6))
    plt.plot(series, label="Data")
    moving_avg = series.rolling(window=12).mean()
    plt.plot(moving_avg, label='Rolling mean')
    moving_sd = series.rolling(window=12).std()
    plt.plot(moving_sd, label='Rolling SD')
    plt.legend(loc='best')
    plt.title('Data with Rolling Mean & Standard Deviation', fontsize=14)
    plt.show()

    dftest = adfuller(series, autolag='AIC')
    print('Results of Dickey-Fuller Test:')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

Let's perform an initial stationarity check:

```python
validate_stationarity(df.sales)
```

So, the time series is apparently not stationary, for several reasons:

1. the standard variation is clearly not constant
2. there is a clear increasing trend, so in particular a non-constant mean
3. there is an obvious seasonal pattern

Before addressing these issues, statsmodels allows for a nice overview of the series' decomposition:

```python
decomp = sm.tsa.seasonal_decompose(df)
decomp.plot()
plt.show()
```

Before going forward, I want to show a plot of the autocorrelation function of the original data, confirming the seasonal pattern with a period of 12 months.

```python
pd.plotting.autocorrelation_plot(df.sales)
plt.show()
```

As a first step towards stationarity, I use a log-transformation to linearize growth, i.e. to make standard deviation more consistent.

```python
df['log_sales'] = np.log(df.sales)
validate_stationarity(df['log_sales'])
```

Next, I use seasonal differencing to get rid of the seasonality.

```python
df['log_sales_sdiff'] = df['log_sales'].diff(periods=12)
validate_stationarity(df['log_sales_sdiff'].dropna())
```

This looks good so far! But even though the Dickey-Fuller Test already confirms stationarity with a confidence level of 5%, I have not yet dealt with the increasing trend of the series. This can be done by performing first order differencing on top of the seasonal differencing.

```python
df['log_sales_sdiff_firstdiff'] = df['log_sales_sdiff'].diff()
validate_stationarity(df['log_sales_sdiff_firstdiff'].dropna())
```

As you can see by the very low p-value, the time series can now definitely be considered stationary!

## Finding good SARIMA parameters

After making the data stationary, I want to find good parameters for a SARIMA forcasting model. This can be done by looking at the autocorrelation function (ACF) and the partial autocorrelation function (PACF). Again, the statsmodels library makes this very easy.

For anyone interested, detailed information about the SARIMA model and its parameters can be found [here](https://www.otexts.org/fpp/8/9).

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df['log_sales_sdiff_firstdiff'].dropna(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df['log_sales_sdiff_firstdiff'].dropna(), lags=40, ax=ax2)
```

Due to the negative autocorrelation in lag 1, and a significant negative partial autocorrelation in lags 1 and 2, an MA component of 1 or 2 seems advisable. In addition, the negative autocorrelation in lag 12 clearly suggests and SMA component of 1. Since I have differenced and seasonally differenced exactly once each, the I and SI components are already given as 1. Together with the known period of 12, this gives the following two SARIMA models:

1. SARIMA(0,1,1)(0,1,1)12
2. SARIMA(0,1,2)(0,1,1)12

## Building and evaluating the models

So, let's build the two models! To do this, I will hold out the last 12 months of the data as test set and use everything else as training set.

**Note:** While the SARIMA model can handle the differencing, it cannot handle transformations such as the logarithmic transformation that I performed. Therefore, I have to use the log_sales data for fitting the model and any predictions have to be converted back via *exp* to match the data's original scale. 

```python
train, test = df.log_sales[:'2007-12'], df.sales['2008-01':]

model1 = sm.tsa.statespace.SARIMAX(train, order=(0,1,1), seasonal_order=(0,1,1,12))
results1 = model1.fit()
results1.summary()
```

```python
model2 = sm.tsa.statespace.SARIMAX(train, order=(0,1,2), seasonal_order=(0,1,1,12))
results2 = model2.fit()
results2.summary()
```

In the summaries above, the *coef* column shows the weight (i.e. importance) of each feature and how each one impacts the time series. The *P>|z|* column informs about the significance of each feature's weight. It can be seen that the second MA term definitely did not yield a significant contributions to the second model's performance, so it is not reasonable to keep it. For this reason, I will continue to use the first model only.

Now, it is important to run model diagnostics to ensure that none of the assumptions made by the model have been violated. The plot_diagnostics object allows to quickly generate model diagnostics and check for any unusual behavior.

```python
results1.plot_diagnostics(figsize=(15, 12))
plt.show()
```

This looks pretty good! One main concern is to ensure that the residuals of a model are uncorrelated and normally distributed with zero-mean. The plots above generally indicate that this is roughly the case here. There might still be room for improvement, but this is out of scope for me at this point.

So, let's look at how well the model is able to predict the test data:

```python
df['forecast_1_log'] = results1.predict(start=df.index.get_loc('2008-01-01'), end=df.index.get_loc('2008-12-01'))
df['forecast_1'] = np.exp(df['forecast_1_log'])
df[['sales', 'forecast_1']].plot(figsize=(12,6))
plt.show()
```

Not too bad! Even though there are notable residuals, the general trend and seasonality of the time series appears to be caught pretty well.

Finally, let's look at some numeric metrics:

```python
from sklearn.metrics import mean_absolute_error, explained_variance_score

print('Mean error:', (df['forecast_1'].dropna() - test).mean())
print('Mean absolute error:', mean_absolute_error(test, df['forecast_1'].dropna()))
print('Explained variance score:', explained_variance_score(test, df['forecast_1'].dropna()))
```

## Alternate SARIMA model with automated parameter selection

As an alternative to my analysis to find appropriate parameters above, it is also possible to use brute force (of course only up to a certain extent) to detect good SARIMA parameters. In the context of statistics and machine learning, such a process is known as *grid search*.

To score models, I will use the Akaike information criterion (AIC). The AIC is an estimator of the relative quality of statistical models for a given set of data, which allows for comparison among multiple SARIMA models with different parameters. The AIC measures how well a model fits the given training data and also takes into account its complexity, i.e. a model using lots of features will be assigned a larger AIC score than a model that uses fewer features to achieve the same goodness-of-fit. 

Therefore, I am interested in finding the model that yields the lowest AIC value. As a baseline, here are the AIC scores of the two SARIMA models from above:

```python
print(results1.aic, results2.aic)
```

Now, I will try several parameter combinations and store the AIC score of each parameter combination to find the *best model* eventually. Because some parameter combinations may lead to numerical misspecifications, I am disabling warning messages in order to avoid an overload of warning messages. These misspecifications can also lead to errors and throw an exception, so I catch these exceptions and ignore the parameter combinations that cause these issues.

```python
from itertools import product
warnings.filterwarnings('ignore')

p = q = range(3) 
pdq = list(product(p, [1], q))
seasonal_pdq = list(product(p, [1], q, [12]))
aic_scores = []

for i in pdq:
    for j in seasonal_pdq:
        try:
            model = sm.tsa.statespace.SARIMAX(train, order=i, seasonal_order=j)
            results = model.fit()
            if results.aic != np.nan:
                aic_scores.append((i, j, results.aic))
        except: # some parameter combinations will not be valid
            pass
```

```python
aic_scores.sort(key=lambda x: x[2])
print('Top 5 parameters (according to AIC scores):')
for t in aic_scores[:5]:
    print('SARIMA{}{} - AIC: {}'.format(*t))
```

Alright! My choice from above makes a good second place. The model with the best AIC only trades the SMA term against an SAR term. Let's check how this model performs in comparison:

```python
best_model = sm.tsa.statespace.SARIMAX(train, order=(0, 1, 1), seasonal_order=(1, 1, 0, 12))
results = best_model.fit()
df['forecast_best_log'] = results.predict(start=df.index.get_loc('2008-01-01'), end=df.index.get_loc('2008-12-01'))
df['forecast_best'] = np.exp(df['forecast_best_log'])
print('AIC:', results.aic)
print('Mean error (best model):', (df['forecast_best'].dropna() - test).mean())
print('Mean absolute error (best model):', mean_absolute_error(test, df['forecast_best'].dropna()))
print('Explained variance score (best model):', explained_variance_score(test, df['forecast_best'].dropna()))
```
